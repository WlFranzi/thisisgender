```python
import pandas as pd
import numpy as np

df = pd.read_csv("./data-raw/liwc-catSum-per-song.csv")
```

```python
def get_gender(value_she, value_he):
    print(value_she, value_he)
    if value_she == 0 and value_he > 0:
        print("male")
        return "male"
    if value_she > 0 and value_he == 0:
        print("female")
        return "female"
    else:
        print("Neutral")
        return "neutral"
```

```python
df_gender = pd.DataFrame()

df_gender["Lyrics"] = df["Lyrics"]
df_gender["Year"] = df["Year"]
df_gender["dominant_gender"] = df.apply(lambda row: get_gender(row["She"], row["He"]), axis=1)
print(df_gender.head())
df_gender.to_csv("./data-label/songs-gender-lable.csv")
```

```python
df_gender = df_gender[df_gender.dominant_gender != "neutral"]
df_gender.to_csv("./data-labled/songs-femal-male-lable.csv")
```
