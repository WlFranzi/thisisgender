```python
import requests

def analyze_tone(text):
    username='blabla'
    password = 'Hakkunamattattaaaaa'
    watsonUrl = 'https://gateway.watsonplatform.net/tone-analyzer/api/v3/tone?version=2017-09-26'
    headers = {"content-type": "text/plain"}
    data = text
    try:
        r = requests.post(watsonUrl, auth=(username,password),headers = headers,
         data=data)
        try:
            print("hey")
            return eval(r.text)['document_tone']['tones'][0]
        except Exception as e:
            print(e)
            return None
    except Exception as e:
        print(e)
        return False

    
import pandas as pd
df = pd.read_csv('./data-labled/songs-femal-male-lable.csv')
print(df.head())
df['Emotions'] = df["Lyrics"].apply(lambda x: analyze_tone(x)) 
```
